### Load MedQuad dataset from Huggingface
(https://huggingface.co/datasets/keivalya/MedQuad-MedicalQnADataset)

```python
from datasets import load_dataset

# Load the MedQuad dataset
ds = load_dataset("keivalya/MedQuad-MedicalQnADataset")

# Inspect splits
print(ds)

# Example: look at the first few examples
print(ds["train"][:5])
```

```python
# 2. Imports
from datasets import load_dataset
from unsloth import FastLanguageModel
import torch
from transformers import TrainingArguments, Trainer, DataCollatorForLanguageModeling
```

```python
# 3. Load dataset
full_dataset = load_dataset("keivalya/MedQuad-MedicalQnADataset")

print(full_dataset)
print(full_dataset["train"][0])


### use 50% data only to speed up finetuning
dataset = full_dataset["train"].train_test_split(
    test_size=0.5,
    seed=42
)

# dataset["train"] = dataset_small["train"]
```

```python
def format_example(example):
    instruction = "Answer the medical question using the provided context."
    input_text = f"Context: {example['qtype']}\n\nQuestion: {example['Question']}"
    output_text = example["Answer"]

    prompt = f"""### Instruction:
                  {instruction}

                  ### Input:
                  {input_text}

                  ### Response:
                  {output_text}"""

    return {"text": prompt}

dataset = dataset.map(format_example)
```

```python
# 5. Load TinyLLaMA with Unsloth QLoRA
max_seq_length = 2048
lora_rank = 16

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "TinyLLaMA/TinyLLaMA-1.1B-Chat-v1.0",
    load_in_4bit = True,
    max_seq_length = max_seq_length,
    fast_inference = False,         # vLLM compatibility
    max_lora_rank = lora_rank,
    use_gradient_checkpointing = "unsloth",
)

tokenizer.pad_token = tokenizer.eos_token

model = FastLanguageModel.get_peft_model(
    model,
    r = lora_rank,
    lora_alpha = 16,
    lora_dropout = 0.05,
    target_modules = ["q_proj","k_proj","v_proj","o_proj","gate_proj","up_proj","down_proj"]
)
```

```python
# 6. Tokenization
def tokenize(batch):
    return tokenizer(
        batch["text"],
        truncation=True,
        max_length=max_seq_length,
        padding="max_length",
    )

tokenized = dataset["train"].map(tokenize, batched=True, remove_columns=dataset["train"].column_names)
```

```python
# 7. Data Collator
data_collator = DataCollatorForLanguageModeling(
    tokenizer,
    mlm=False,
)
```

```python
# 8. Training arguments
training_args = TrainingArguments(
    output_dir="./tinyllama-medquad",
    per_device_train_batch_size=2,
    gradient_accumulation_steps=8,
    warmup_steps=50,
    max_steps=1500,                       # or use num_train_epochs=2
    learning_rate=2e-4,
    fp16=True,
    logging_steps=25,
    save_steps=250,
    optim="paged_adamw_8bit",
    report_to="none",
)
```

```python
# 9. Train & save model
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized,
    data_collator=data_collator,
)

trainer.train()

model.save_pretrained("./tinyllama-medquad-lora")
tokenizer.save_pretrained("./tinyllama-medquad-lora")

print("Training complete! Model saved to tinyllama-medquad-lora")
```
